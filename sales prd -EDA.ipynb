```python
data_sum_by_Year= dfw[['Year','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].groupby(['Year']).sum().reset_index()
data_sum_by_Year.sort_values(by='Year')
```

```python
x=data_sum_by_Year['Year']
y1=data_sum_by_Year["NA_Sales"]
y2=data_sum_by_Year["EU_Sales"]
y3=data_sum_by_Year["JP_Sales"]
y4=data_sum_by_Year["Other_Sales"]
y5=data_sum_by_Year["Global_Sales"]
plt.plot(x,y1,label="NA_Sales")
plt.plot(x,y2,label="EU_Sales")
plt.plot(x,y3,label="JP_Sales")
plt.plot(x,y4,label="Other_Sales")
plt.plot(x,y5,label="Global_Sales")
plt.xlabel("Year",fontfamily="Times New Roman",fontsize=18)
plt.ylabel("sale",fontfamily="Arial",fontsize=18)
plt.title('COMPARE SALES',fontfamily="Arial",fontsize=22)
plt.legend()
plt.grid(True)
plt.show()
```
